```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Plot data
df = pd.read_csv('./data/cereal.csv')
df.head()
```

```python
#Eliminate irrelevant columns
cols = ['shelf', 'weight', 'cups', 'rating', 'mfr', 'type']
df = df.drop(columns = cols)
df.head()
```

```python
#Get correlation data
corr_table = df.corr()
corr_table
```

```python
#Build matrix booleans
x = np.ones_like(corr_table, dtype = 'bool')
x
```

```python
#Return only upper triangle matrix
y = np.triu(x)
y
```

```python
sns.heatmap(data = corr_table, mask = y);
```

```python
adjusted_y = y[1:, :-1]
adjusted_corr_table = corr_table.iloc[1:, :-1]
cmap = sns.diverging_palette(60,0,90,50, as_cmap = True)
fig, ax = plt.subplots(figsize=(6, 5))
ax.set(title = 'Correlation Matrix of Nutrients in Cereal');
sns.heatmap(data = adjusted_corr_table, mask = adjusted_y, cmap = cmap, vmin = -1, vmax = 1,
            annot = True, fmt = '.3f');
```
